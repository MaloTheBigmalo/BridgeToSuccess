```python
Auto = pd.read_csv('/content/drive/MyDrive/data/Auto.data',
                    na_values=['?'],
                    sep=r"\s+")
```

```python
Auto.shape # Auto.shape attribute tells us that the data has 397 observations or rows and 9 columns.
```

```python
num_rows = Auto.isna().any(axis=1).sum() # This count the rows with missing values and we print it nextly.
print(f"Number of rows with missing values: {num_rows}")
```

```python
# Count rows that contain only missing values
number = Auto.isna().all(axis=1).sum()
print(number)
```

```python
# Get the actual rows with missing values
rows_with_missing_values = Auto[Auto.isna().any(axis=1)]
print(rows_with_missing_values)
```

```python
#Since we only have 5 of the rows containing missing values, we can remove them using Auto.dropna methode
Auto_new = Auto.dropna()
Auto_new.shape
```

```python
# Override Auto
Auto = Auto_new
```
